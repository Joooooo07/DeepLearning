```python
# Exercise 1 - OR Gate using Perceptron

class Perceptron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias

    def step_function(self, x):
        if x >= 0:
            return 1
        else:
            return 0

    def predict(self, inputs):
        # Calculate net input
        net_input = 0
        for i in range(len(inputs)):
            net_input += inputs[i] * self.weights[i]

        net_input += self.bias

        # Apply activation function
        return self.step_function(net_input)


# Suitable weights and bias for OR gate
weights = [1, 1]
bias = -0.5

# Create perceptron
p = Perceptron(weights, bias)

# OR Gate Inputs
inputs = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

print("X1 X2 Output")
for inp in inputs:
    output = p.predict(inp)
    print(inp[0], " ", inp[1], "   ", output)
```

```python
# Exercise 3 - NOR Gate using Perceptron

import matplotlib.pyplot as plt
import numpy as np

class Perceptron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias

    def step_function(self, x):
        if x >= 0:
            return 1
        else:
            return 0

    def predict(self, inputs):
        net_input = 0
        for i in range(len(inputs)):
            net_input += inputs[i] * self.weights[i]

        net_input += self.bias

        return self.step_function(net_input)


# Suitable weights and bias for NOR gate
weights = [-1, -1]
bias = 0.5

# Create perceptron
p = Perceptron(weights, bias)

# NOR Gate Inputs
inputs = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

print("X1 X2 Output")
for inp in inputs:
    output = p.predict(inp)
    print(inp[0], " ", inp[1], "   ", output)

# -------------------------
# Decision Boundary
# -------------------------

x = np.linspace(-0.5, 1.5, 100)

# Decision Boundary:
# -x1 - x2 + 0.5 = 0
y = 0.5 - x

plt.figure(figsize=(5,5))

plt.plot(x, y, label='Decision Boundary')

# Plot input points
for point in inputs:
    if p.predict(point) == 1:
        plt.scatter(point[0], point[1], color='green', s=100)
    else:
        plt.scatter(point[0], point[1], color='red', s=100)

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("NOR Gate Decision Boundary")
plt.grid(True)
plt.legend()

plt.show()
```

```python
# Exercise 4 - Find Suitable Weights

class Perceptron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias

    def step_function(self, x):
        return 1 if x >= 0 else 0

    def predict(self, inputs):
        net_input = sum(i * w for i, w in zip(inputs, self.weights))
        net_input += self.bias
        return self.step_function(net_input)


# Input combinations
inputs = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

# Suitable weights
gates = {
    "AND": [0.5, 0.5],
    "OR": [1, 1],
    "NAND": [-1, -1]
}

bias = -0.5

for gate, weights in gates.items():
    print("\n", gate, "Gate")
    print("X1 X2 Output")

    # Change bias for NAND because -0.5 cannot implement NAND correctly
    if gate == "NAND":
        p = Perceptron(weights, 1.5)
    else:
        p = Perceptron(weights, bias)

    for inp in inputs:
        print(inp[0], " ", inp[1], "   ", p.predict(inp))
```

```python
# Exercise 5 - Activation Function Comparison

import math

# Input combinations
inputs = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

# Weights and Bias for OR Gate
weights = [1, 1]
bias = -0.5

# Step Function
def step(x):
    if x >= 0:
        return 1
    else:
        return 0

# Sigmoid Function
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

# ReLU Function
def relu(x):
    return max(0, x)

print("X1 X2 Net Step Sigmoid ReLU")

for inp in inputs:
    net = inp[0] * weights[0] + inp[1] * weights[1] + bias

    print(
        inp[0],
        " ",
        inp[1],
        " ",
        round(net, 2),
        " ",
        step(net),
        "    ",
        round(sigmoid(net), 3),
        "    ",
        round(relu(net), 3)
    )
```

# Answers

## 1. Which activation is suitable for logic gates?

**Answer:**

The **Step Function** is the most suitable activation function for implementing logic gates because it produces binary outputs (0 or 1), which match the output of logic gates.

## 2. Why?

- **Step Function:** Produces only **0 or 1**, exactly like logic gates.
- **Sigmoid:** Produces values between **0 and 1**, so it does not directly give binary outputs.
- **ReLU:** Produces values from **0 to infinity**, which does not match the binary output of logic gates.
