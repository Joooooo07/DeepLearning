```python
# Exercise 6 - Weight Updation (One Epoch)

# Training data for AND Gate
inputs = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

targets = [0, 0, 0, 1]

# Initial parameters
weights = [0.3, 0.1]
bias = 0
learning_rate = 0.2

# Step activation function
def step(net):
    if net >= 0:
        return 1
    else:
        return 0

print("One Epoch Training\n")

print("X1 X2 Target Net Output Error Updated Weights Updated Bias")

for x, target in zip(inputs, targets):

    # Calculate net input
    net = x[0] * weights[0] + x[1] * weights[1] + bias

    # Predict output
    output = step(net)

    # Calculate error
    error = target - output

    # Update weights
    weights[0] = weights[0] + learning_rate * error * x[0]
    weights[1] = weights[1] + learning_rate * error * x[1]

    # Update bias
    bias = bias + learning_rate * error

    print(x[0], x[1], "   ",
          target, "   ",
          round(net, 2), "   ",
          output, "   ",
          error, "   ",
          [round(weights[0], 2), round(weights[1], 2)],
          "   ",
          round(bias, 2))
```

```python
# Exercise 7 - Epoch Analysis

# Training data for AND Gate
inputs = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

targets = [0, 0, 0, 1]

learning_rate = 0.2

# Step activation function
def step(net):
    return 1 if net >= 0 else 0

# Function to train perceptron
def train(epochs):
    weights = [0.3, 0.1]
    bias = 0

    print(f"\nTraining for {epochs} Epochs")
    print("Epoch\tTotal Error")

    for epoch in range(1, epochs + 1):
        total_error = 0

        for x, target in zip(inputs, targets):

            net = x[0] * weights[0] + x[1] * weights[1] + bias
            output = step(net)

            error = target - output
            total_error += abs(error)

            # Update weights
            weights[0] += learning_rate * error * x[0]
            weights[1] += learning_rate * error * x[1]

            # Update bias
            bias += learning_rate * error

        print(epoch, "\t", total_error)

        # Stop if converged
        if total_error == 0:
            print("Perceptron converged at Epoch", epoch)
            break


# Train for different epochs
train(5)
train(10)
train(20)
```

The perceptron converges when the total error becomes 0. In this example, it converges at Epoch 4.

```python
# Exercise 8 - Learning Rate Comparison

# Training data for AND Gate
inputs = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

targets = [0, 0, 0, 1]

# Step activation function
def step(net):
    return 1 if net >= 0 else 0

# Function to train perceptron
def train(learning_rate):
    weights = [0.3, 0.1]
    bias = 0
    epochs = 0

    while True:
        total_error = 0
        epochs += 1

        for x, target in zip(inputs, targets):

            net = x[0] * weights[0] + x[1] * weights[1] + bias
            output = step(net)

            error = target - output
            total_error += abs(error)

            # Update weights
            weights[0] += learning_rate * error * x[0]
            weights[1] += learning_rate * error * x[1]

            # Update bias
            bias += learning_rate * error

        if total_error == 0:
            break

    # Calculate accuracy
    correct = 0
    for x, target in zip(inputs, targets):
        net = x[0] * weights[0] + x[1] * weights[1] + bias
        output = step(net)
        if output == target:
            correct += 1

    accuracy = (correct / len(targets)) * 100

    return epochs, weights, accuracy


learning_rates = [0.1, 0.5, 1.0]

print("Learning Rate Comparison")
print("-" * 55)
print("Learning Rate\tEpochs\tFinal Weights\t\tAccuracy")

for lr in learning_rates:
    epochs, weights, accuracy = train(lr)
    print(f"{lr}\t\t{epochs}\t{weights}\t{accuracy:.2f}%")
```

```python
# Exercise 9 - XOR Gate using Single Perceptron

class Perceptron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias

    def step_function(self, x):
        if x >= 0:
            return 1
        else:
            return 0

    def predict(self, inputs):
        net = 0
        for i in range(len(inputs)):
            net += inputs[i] * self.weights[i]

        net += self.bias

        return self.step_function(net)


# Weights and bias
weights = [1, 1]
bias = -0.5

# Create perceptron
p = Perceptron(weights, bias)

# XOR Gate Inputs
inputs = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

print("X1 X2 Predicted Output")

for inp in inputs:
    output = p.predict(inp)
    print(inp[0], " ", inp[1], "      ", output)

print("\nExpected XOR Output")
print("0 0 -> 0")
print("0 1 -> 1")
print("1 0 -> 1")
print("1 1 -> 0")
```

1. Does it work?

Answer:

No. A single perceptron cannot correctly implement the XOR gate.

2. Explain why.

Answer:

The XOR gate is not linearly separable. A single perceptron can classify only linearly separable data. Since the XOR inputs cannot be separated by a single straight line, one perceptron cannot learn the XOR function. Therefore, a Multi-Layer Perceptron (MLP) with a hidden layer is required.

```python
# Exercise 10 - XOR using Multi-Layer Perceptron

import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# XOR Input
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# XOR Output
y = np.array([
    [0],
    [1],
    [1],
    [0]
])

# Build MLP Model
model = Sequential([
    Dense(2, input_dim=2, activation='sigmoid'),
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Train the model
model.fit(X, y, epochs=500, verbose=0)

# Predict outputs
predictions = model.predict(X)

print("X1 X2 Predicted Output")

for i in range(len(X)):
    output = 1 if predictions[i][0] >= 0.5 else 0
    print(X[i][0], " ", X[i][1], "      ", output)
```

Result

The Multi-Layer Perceptron with one hidden layer and two hidden neurons successfully learns the XOR gate and produces the correct outputs.
